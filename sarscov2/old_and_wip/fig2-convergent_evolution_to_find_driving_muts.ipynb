Mutations that appear many times over course of natural evoltion of SARS-CoV-2 and are inherited by a descending clade (don't just die out) could be beneficial. They are likely neutral at worst. Try to look for mutations driving clade success by finding mutations that occur repeatedly and result in high logisitic growth rates and high numbers of S1 mutations each time

Start by finding the number of independent times each mutation occurs and doesn't die out (forms a resulting clade of at least 10 tips)

*allow all analyses to be done for specific amino acid mutation or by position. With `mutation_type` = "aa" or "site"

```python
#Download tree json
# tree_url = "https://data.nextstrain.org/ncov_global_2021-05-20.json"
tree_url = "https://data.nextstrain.org/ncov_global.json"

tree_json = requests.get(tree_url).json()

#Put tree in Bio.Phylo format
tree = json_to_tree(tree_json)
```

Read in reference sequence 

```python
# make dictionary with gene name as key and reference sequence of that gene as value

reference_sequence = {}

for record in SeqIO.parse(open("reference_seq_edited.gb","r"), "genbank"):
    for feature in record.features:
        if feature.type == 'CDS':
            gene_seq = feature.location.extract(record.seq).translate()
            reference_sequence[feature.qualifiers['gene'][0]] = gene_seq
```

```python
def count_independent_occurrences(mutation_type):
    
    all_mutations = []


    #only look at mutations on internal branches
    for node in tree.find_clades(terminal=False):
        
        # only consider mutations on branches that give rise to a clade of at least 10 tips
        if len(node.get_terminals()) >=10:

            if hasattr(node, 'branch_attrs'):
                for gene, mut_list in node.branch_attrs["mutations"].items():
                    # not considering synonymous mutations for now
                    if gene!= 'nuc':
                        for mut in mut_list:
                            
                            if mutation_type == 'aa':
                                # look at specific nonsyn muts
                                gene_mutation = f'{gene}_{mut}'

                                
                            elif mutation_type == 'site':
                                # look at mutation to a codon (without regard to identity of mutation)
                                gene_mutation = f'{gene}_{mut[1:-1]}'
                            
                            # append mutation to a list of all observed mutations
                            # exclude if mutation is in stop codon position of gene
                            if int(mut[1:-1])!=len(reference_sequence[gene]):
                                all_mutations.append(gene_mutation)
                




    all_mutations_unique = list(set(all_mutations))    

    independent_occurrences_all_mutations = Counter(all_mutations)
    

    return all_mutations_unique, independent_occurrences_all_mutations
```

For each clade: 
1. find the number of S1 muts and the logistic growth rate
2. find all mutations that happened on path from root to clade
3. for each mutation in (2), add the values from (1) to a list. 

Over all clades, the result will be a dictionary where each observed mutation is a key and the values are the logistic growth rates of all clades that descend from the observed mutation. 

```python
#Function to find path from root to clade
def get_parent(tree, child_clade):
    node_path = tree.get_path(child_clade)
    return node_path
```

```python
def clades_descending_from_mut(mutation_type, all_mutations_unique):

    #initialize dict for storing all growth rates observed in clades descending from {key} mutation
    growth_by_precedingmut_dict = {k:[] for k in all_mutations_unique}


    #initialize dict for storing S1 mutation counts observed in clades descending from {key} mutation
    s1muts_by_precedingmut_dict = {k:[] for k in all_mutations_unique}



    for node in tree.find_clades():
        
        # still only considering branches that give rise to a clade of at least 10
        if len(node.get_terminals())>=10:
        
            node_path = get_parent(tree, node)


            s1_mutations = None
            if "S1_mutations" in node.node_attrs:
                s1_mutations = node.node_attrs["S1_mutations"]["value"]


            logistic_growth = None
            if "logistic_growth" in node.node_attrs:
                logistic_growth = node.node_attrs["logistic_growth"]["value"]


            for parent in node_path[:-1]:
                if hasattr(parent, 'branch_attrs'):
                    for gene, mut_list in parent.branch_attrs["mutations"].items():
                        if gene!= 'nuc':
                            for mut in mut_list:

                                if mutation_type == 'aa':
                                    # look at specific nonsyn muts
                                    gene_mutation = f'{gene}_{mut}'


                                elif mutation_type == 'site':
                                    # look at mutation to a codon (without regard to identity of mutation)
                                    gene_mutation = f'{gene}_{mut[1:-1]}'


                                # exclude mutations to last codon of gene (stop codon)
                                if int(mut[1:-1])!=len(reference_sequence[gene]):

                                    if logistic_growth!=None:
                                        growth_by_precedingmut_dict[gene_mutation].append(logistic_growth)

                                    if s1_mutations!=None:
                                        s1muts_by_precedingmut_dict[gene_mutation].append(s1_mutations)
                                        
    # get rid of empty entries
    growth_by_precedingmut_dict = {k:v for k,v in growth_by_precedingmut_dict.items() if v}
    s1muts_by_precedingmut_dict = {k:v for k,v in s1muts_by_precedingmut_dict.items() if v}
    
    
    return growth_by_precedingmut_dict, s1muts_by_precedingmut_dict
```

Find the average logisitic growth rate after a mutation by finding the mean grwoth rate of all clades descending from  branches with that mutation. Do the same for number of S1 mutations. 

Then, make a dataframe where each row is a mutation and columns contain number of independent occurrences, average growth rates, and number of S1 mutations.

```python
def calc_mean_rate_and_s1_muts(growth_by_precedingmut_dict, s1muts_by_precedingmut_dict):
    
    # to consider: clade nesting-> just take unique values? 
    # Here, all nested clades are averaging, giving more weight to larger clades
    growth_by_precedingmut_mean = {k:(sum(v)/len(v)) for k,v in growth_by_precedingmut_dict.items()}

    # Here, only unique values are taken, to limit the effect of nesting clades
    growth_by_precedingmut_limit_nesting = {k:list(set(v)) for k,v in growth_by_precedingmut_dict.items()}
    growth_by_precedingmut_limit_nesting_mean = {k:(sum(v)/len(v)) for k,v in growth_by_precedingmut_limit_nesting.items()}

    # this will give more weight to larger clades
    s1muts_by_precedingmut_mean = {k:(sum(v)/len(v)) for k,v in s1muts_by_precedingmut_dict.items() if len(v)!=0}

#     return  growth_by_precedingmut_mean, s1muts_by_precedingmut_mean
    return growth_by_precedingmut_limit_nesting_mean, s1muts_by_precedingmut_mean
```

```python
def convergent_evo_dataframe(mutation_type, growth_by_precedingmut_dict, 
                             s1muts_by_precedingmut_dict, independent_occurrences_all_mutations):
    
    growth_by_precedingmut_mean, s1muts_by_precedingmut_mean = calc_mean_rate_and_s1_muts(growth_by_precedingmut_dict, s1muts_by_precedingmut_dict)
    
    # make dataframe where each row is a mutation
    # and columns contain number of independent occurrences, average growth rates, and number of S1 mutations

    convergent_evo_list = []

    for mut, occurrences in independent_occurrences_all_mutations.items():
        if mut in s1muts_by_precedingmut_mean.keys():
            avg_s1_muts = s1muts_by_precedingmut_mean[mut]
        else:
            avg_s1_muts = None

        if mut in growth_by_precedingmut_mean.keys():
            avg_growth = growth_by_precedingmut_mean[mut]
        else:
            avg_growth = None



        # for deletions that span multiple consecutive residues, condense this into one row
        if mut == 'ORF1a_S3675-': 
            mut = 'ORF1a_3675/3676/3677del'


        elif mut == 'S_H69-':
            mut = 'S_69/70del'

        if mut not in ['ORF1a_G3676-', 'ORF1a_F3677-', 'S_V70-']:
            convergent_evo_list.append({'mutation': mut, 'independent_occurrences': occurrences, 
                                        'avg_s1_muts': avg_s1_muts, 'avg_growth': avg_growth
                                       })

    convergent_evo_df = pd.DataFrame(convergent_evo_list)


    # multiply occurrences times growth
    convergent_evo_df['occurrences_times_growth']= convergent_evo_df['independent_occurrences']*convergent_evo_df['avg_growth']

    
    return convergent_evo_df
```

Now, for each mutation identified in figures 2B and 2C, find how many of the independent occurrences of these mutationsn result in an emerging lineage

```python
def location_of_muts(mutation_type, top_scoring_muts):
    
    # find the start position of the codon encoding each site
    genome_location_of_sites = {}


    for mutation in top_scoring_muts:
        gene = mutation.split('_')[0]

        if mutation_type == 'aa':
            if mutation[-3:]!='del':
                mut_pos = mutation.split('_')[1][1:-1]

            elif mutation[-3:]=='del':
                mut_pos = mutation.split('_')[1].split('/')[0]

        elif mutation_type == 'site':
            mut_pos = mutation.split('_')[1]

        for record in SeqIO.parse(open("reference_seq_edited.gb","r"), "genbank"):
            for feature in record.features:
                if feature.type == 'CDS':
                    if feature.qualifiers['gene'][0] == gene:
                        mut_location_start = feature.location.start + (((int(mut_pos)-1)*3)+1)
                        genome_location_of_sites[mut_location_start] = mutation
              
        
    return genome_location_of_sites
```

```python
def genotypes_at_top_sites(genome_location_of_sites):
    
    # initialize dictionary to store genotype info
    # key is name of isolate, values are dictionary of genotype at each of the top mutation sites
    tip_genotypes = {}

    # go through all sequences isolates
    for k,v in record_dict.items():
        this_tip = {}
        # store genotype at each mutation site
        for nt_start, mutation in genome_location_of_sites.items():

            nt_codon = v.seq[nt_start-1:nt_start+2]
            if all([characters in ['A', 'C', 'G', 'T'] for characters in nt_codon]):
                this_tip[mutation] = str(nt_codon.translate())
            elif all([characters in ['-'] for characters in nt_codon]):
                this_tip[mutation] = '-'
            else:
                this_tip[mutation] = None

        tip_genotypes[k] = this_tip

    return tip_genotypes
```

```python
def genotype_prevelance_in_emerging_lineages(mutation_type, top_scoring_muts, tip_genotypes, genome_location_of_sites):

    #initiate list to store all clade info
    clade_stats = []

    emerging_lineage_tips = {}

    # don't want to double count, so only look at base of emerging lineage clade
    clade_bases = []

    for node in tree.find_clades(terminal=False):
        
        if 'logistic_growth' in node.node_attrs:

            if "emerging_lineage" in node.node_attrs:
                emerging_lineage = node.node_attrs["emerging_lineage"]["value"]

                if emerging_lineage!='unassigned':

                    clade_already_visited = False

                    node_path = get_parent(tree, node)
                    for parent in node_path:
                        if parent in clade_bases:
                            clade_already_visited = True

                    if clade_already_visited == False:
            #                     print(node.__dict__)
                        clade_bases.append(node)

                        logistic_growth = node.node_attrs["logistic_growth"]["value"]

                        tips_in_clade = node.get_terminals()
                        tip_names_in_clade = [t.name for t in tips_in_clade]

                        # tally the number of tips with given mutation within the clade
                        clade_genotype_tally = {k:0 for k in top_scoring_muts}
                        clade_genotype_tally_denominator = {k:0 for k in top_scoring_muts}
                        for tip in tip_names_in_clade:
                            for mutation, aa in tip_genotypes[tip].items():
                                if aa!= None:
                                    clade_genotype_tally_denominator[mutation]+=1

                                    # for `aa`, want to see fraction of emerging lineage with this exact genotype
                                    if mutation_type == 'aa':
                                        if mutation[-3:]=='del':
                                            if aa == '-':
                                                clade_genotype_tally[mutation]+=1

                                        if aa == mutation[-1]:
                                            clade_genotype_tally[mutation]+=1


                                    # for `site`, want to see whether position is different than reference seq
                                    elif mutation_type == 'site':
                                        # all deletions will be different from reference sequence
                                        if aa == '-':
                                            clade_genotype_tally[mutation]+=1
                                        else:
                                            codon_start= list(genome_location_of_sites.keys()
                                                             )[list(genome_location_of_sites.values()
                                                                   ).index(mutation)]
                                            for record in SeqIO.parse(open("reference_seq_edited.gb","r"), "genbank"):
                                                ref_aa = record.seq[codon_start-1:codon_start+2].translate()

                                            if aa != ref_aa:
                                                clade_genotype_tally[mutation]+=1


                        # find proportion of tips that have the given mutation
                        clade_genotype_freqs = {}
                        for k, v in clade_genotype_tally.items():
                            freq = v/clade_genotype_tally_denominator[k]

                            clade_genotype_freqs[k] = freq


                        clade_stats.append({'emerging_lineage': emerging_lineage, 
                                            'logistic_growth': logistic_growth,
                                            **clade_genotype_freqs})


    clade_stats_df = pd.DataFrame(clade_stats).fillna(0)
    
    return clade_stats_df
```

```python
def fig2e_linear_regression(mutation_type, filename = False):
    
    all_mutations_unique, independent_occurrences_all_mutations = count_independent_occurrences(mutation_type)

    top_scoring_muts_s1 = fig2c_occurrences_vs_s1muts(mutation_type, plot = False)
    top_scoring_muts_growth = fig2b_occurrences_vs_growth(mutation_type, plot = False)
    top_scoring_muts = list(set(top_scoring_muts_s1+top_scoring_muts_growth))

    genome_location_of_sites = location_of_muts(mutation_type, top_scoring_muts)
    tip_genotypes = genotypes_at_top_sites(genome_location_of_sites)


    #initiate list to store all clade info
    clade_stats_linear_reg = []

    #Only want to look at clades, don't care about tips
    for node in tree.find_clades(terminal=False):
        
        if len(node.get_terminals())>=10:

            # only care about clade if logistic growth value is associated
            # find stored logistic growth value
            if "logistic_growth" in node.node_attrs:
                logistic_growth = node.node_attrs["logistic_growth"]["value"]

                #all S1 muts from root to clade already stored as json value
                if "S1_mutations" in node.node_attrs:
                    s1_mutations = node.node_attrs["S1_mutations"]["value"]
                else:
                    s1_mutations = None

                # find all tips in this clade
                tips_in_clade = node.get_terminals()
                tip_names_in_clade = [tip.name for tip in tips_in_clade]

                # tally the genotype of each tip within the clade
                clade_mutation_tally = {x:0 for x in top_scoring_muts}
                clade_total_tally = {x:0 for x in top_scoring_muts}
                for tip in tip_names_in_clade:
                    for mutation, aa in tip_genotypes[tip].items():

                        # ignore the isolates that were not sequenced
                        if aa != None:
                            #count the total number of tips sequenced at this site
                            clade_total_tally[mutation] +=1
                            # tally if this tip has the mutation
                            if mutation[-3:]=='del':
                                if aa == '-':
                                    clade_mutation_tally[mutation] +=1
                            if aa == mutation[-1]:
                                clade_mutation_tally[mutation] +=1



                # find proportion of tips that have given genotype
                clade_genotype_freqs = {}
                for k, count in clade_mutation_tally.items():
                    total_sequenced_tips = clade_total_tally[k]
                    if total_sequenced_tips == 0:
                        freq = None
                    else:
                        freq = count / total_sequenced_tips
                    clade_genotype_freqs[k] = freq


                clade_stats_linear_reg.append({'clade': node.name, 'logistic_growth': logistic_growth, 
                                    'num_s1_muts': s1_mutations, **clade_genotype_freqs})


    clade_stats_linear_reg_df = pd.DataFrame(clade_stats_linear_reg).fillna(0)
    
#     clade_stats_linear_reg_df.to_csv('dataframe.tsv', sep='\t')
    
    print(len(clade_stats_linear_reg_df))
    
    clade_stats_linear_reg_df = clade_stats_linear_reg_df.drop_duplicates(subset=['logistic_growth'])
    
    print(len(clade_stats_linear_reg_df))
    
    # greedy linear regression, where (at each iteration) a new best feature is chosen
    all_mut_keys = list(clade_stats_linear_reg_df.keys())[2:]


    X = clade_stats_linear_reg_df[all_mut_keys]

    y = clade_stats_linear_reg_df['logistic_growth']
    
    lasso = linear_model.LassoCV().fit(X, y)
    
    knn = KNeighborsClassifier(n_neighbors=3)
    sfs = SequentialFeatureSelector(knn, n_features_to_select=3).fit(X,y)
    
    print(f'identified by SFS neighbors {np.array(all_mut_keys)[list(sfs.get_support())]}')
    

    sfs_forward = SequentialFeatureSelector(lasso, n_features_to_select=3,
                                            direction='forward').fit(X, y)

    
    print(f'identified by SFS forward {np.array(all_mut_keys)[list(sfs_forward.get_support())]}')
    

    
    
    importance = np.abs(lasso.coef_)
    feature_names = np.array(all_mut_keys)
    fig, ax = plt.subplots(figsize=(7,2))
    plt.bar(height=importance, x=feature_names, color = '#64B6C4')
    
    ax.spines['top'].set_color('white')
    ax.spines['right'].set_color('white')
    ax.set_ylabel('Lasso model coefficient', size=12)
    
    plt.xticks(rotation = 45, ha='right')


#     ranked_importance_by_index = sorted(range(len(importance)), key=lambda x: importance[x])
#     top_ranked_idicies = reversed(ranked_importance_by_index[-20:])
#     for index in top_ranked_idicies:
#         variable = all_mut_keys[index]
#         i = importance[index]
#         print(variable, i)
        
    if filename:
        fig.savefig(filename, dpi=300, bbox_inches='tight')
        
```

```python
def fig2f_growth_depending_on_constituent_muts(mutation_type):
    
    all_mutations_unique, independent_occurrences_all_mutations = count_independent_occurrences(mutation_type)

    top_scoring_muts_s1 = fig2c_occurrences_vs_s1muts(mutation_type, plot = False)
    top_scoring_muts_growth = fig2b_occurrences_vs_growth(mutation_type, plot = False)
    top_scoring_muts = list(set(top_scoring_muts_s1+top_scoring_muts_growth))

    genome_location_of_sites = location_of_muts(mutation_type, top_scoring_muts)
    tip_genotypes = genotypes_at_top_sites(genome_location_of_sites)


    #initiate list to store all clade info
    clade_mut_freqs = []

    #Only want to look at clades, don't care about tips
    for node in tree.find_clades(terminal=False):
        
        if len(node.get_terminals())>=10:

            # only care about clade if logistic growth value is associated
            # find stored logistic growth value
            if "logistic_growth" in node.node_attrs:
                logistic_growth = node.node_attrs["logistic_growth"]["value"]

                #all S1 muts from root to clade already stored as json value
                if "S1_mutations" in node.node_attrs:
                    s1_mutations = node.node_attrs["S1_mutations"]["value"]
                else:
                    s1_mutations = None

                # find all tips in this clade
                tips_in_clade = node.get_terminals()
                tip_names_in_clade = [tip.name for tip in tips_in_clade]

                # tally the genotype of each tip within the clade
                clade_mutation_tally = {x:0 for x in top_scoring_muts}
                clade_total_tally = {x:0 for x in top_scoring_muts}
                for tip in tip_names_in_clade:
                    for mutation, aa in tip_genotypes[tip].items():

                        # ignore the isolates that were not sequenced
                        if aa != None:
                            #count the total number of tips sequenced at this site
                            clade_total_tally[mutation] +=1
                            # tally if this tip has the mutation
                            if mutation[-3:]=='del':
                                if aa == '-':
                                    clade_mutation_tally[mutation] +=1
                            if aa == mutation[-1]:
                                clade_mutation_tally[mutation] +=1



                # find proportion of tips that have given genotype
                clade_genotype_freqs = {}
                for k, count in clade_mutation_tally.items():
                    total_sequenced_tips = clade_total_tally[k]
                    if total_sequenced_tips == 0:
                        freq = None
                    else:
                        freq = count / total_sequenced_tips
                    clade_genotype_freqs[k] = freq


                clade_mut_freqs.append({'clade': node.name, 'logistic_growth': logistic_growth, 
                                    'num_s1_muts': s1_mutations, **clade_genotype_freqs})


    clade_mut_freqs_df = pd.DataFrame(clade_mut_freqs).fillna(0)
    
#     clade_mut_freqs_df = clade_mut_freqs_df.drop_duplicates(subset=['logistic_growth'])
    
    all_mut_keys = list(clade_mut_freqs_df.keys())[3:]
    
    # initialize dictionary to store average growth rates of clades 
    # that have given mutation present at at least 50% frequency
    individual_mut_growth_rates = {}
    paired_mut_growth_rates = {}
    threeway_mut_growth_rates = {}
    
    for mut1 in all_mut_keys:
        mut1_logistic_growth_avg = clade_mut_freqs_df[clade_mut_freqs_df[mut1]>0.5]['logistic_growth'].mean()
        individual_mut_growth_rates[mut1] = mut1_logistic_growth_avg
        for mut2 in all_mut_keys:
            if mut1!= mut2:
                mut2_logistic_growth_avg = clade_mut_freqs_df[clade_mut_freqs_df[mut2]>0.5]['logistic_growth'].mean()
                
                pair_cooccurrences_count = len(clade_mut_freqs_df[
                    (clade_mut_freqs_df[mut1]>0.5) & (clade_mut_freqs_df[mut2]>0.5)])

                
                if pair_cooccurrences_count>= 2:
                
                    pair_logistic_growth_avg = clade_mut_freqs_df[
                        (clade_mut_freqs_df[mut1]>0.5) & (clade_mut_freqs_df[mut2]>0.5)]['logistic_growth'].mean()

                    paired_rate_expected = mut1_logistic_growth_avg * mut2_logistic_growth_avg
                    
                    pair_observed_expected_ratio = pair_logistic_growth_avg/paired_rate_expected

                    paired_mut_growth_rates[f'{mut1}/{mut2}'] = {'pair_expected': paired_rate_expected, 
                                                                 'pair_rate_actual': pair_logistic_growth_avg, 
                                                                 'pair_ratio': pair_observed_expected_ratio}
                    

            
                for mut3 in all_mut_keys:
                    if mut3 != mut2 and mut3 != mut1:
                        mut3_logistic_growth_avg = clade_mut_freqs_df[clade_mut_freqs_df[mut3]>0.5]['logistic_growth'].mean()
                        
                        threeway_cooccurrences_count = len(clade_mut_freqs_df[
                            (clade_mut_freqs_df[mut1]>0.5) & 
                            (clade_mut_freqs_df[mut2]>0.5) & 
                            (clade_mut_freqs_df[mut3]>0.5)])
                        
                        if threeway_cooccurrences_count >= 2:
                            
                            threeway_rate_expected = mut1_logistic_growth_avg * mut2_logistic_growth_avg * mut3_logistic_growth_avg 

                            
                            threeway_logistic_growth_avg = clade_mut_freqs_df[
                                (clade_mut_freqs_df[mut1]>0.5) & 
                                (clade_mut_freqs_df[mut2]>0.5) & 
                                (clade_mut_freqs_df[mut3]>0.5)]['logistic_growth'].mean()
                            
                            threeway_observed_expected_ratio = threeway_logistic_growth_avg/threeway_rate_expected

                            threeway_mut_growth_rates[f'{mut1}/{mut2}/{mut3}'] = {'threeway_expected': threeway_rate_expected, 
                                                                                  'threeway_rate_actual':threeway_logistic_growth_avg, 
                                                                                  'threeway_ratio': threeway_observed_expected_ratio}

    
#     clade_mut_freqs_df.to_csv('dataframe.tsv', sep='\t')

#     print(individual_mut_frequencies)
#     print(paired_mut_frequencies)
#     print(threeway_mut_frequencies)

    return individual_mut_growth_rates, paired_mut_growth_rates, threeway_mut_growth_rates

        
```
